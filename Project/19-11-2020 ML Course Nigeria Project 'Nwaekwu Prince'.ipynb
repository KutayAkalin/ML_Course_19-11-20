```python
#since the numbers 3 - 8 dont mean anything for quality I would map them to (0)low, (1)mid, (2)high

bins = [0, 5.5, 7.5, 10] #this means 3-5(low), 6-7(mid), 8-9(high)
labels = [0, 1, 2]
df['quality'] = pd.cut(df['quality'], bins=bins, labels=labels)
```

```python
#I chose this columns based on correlations
x= df.loc[:,['fixed acidity', 'volatile acidity','residual sugar',
       'chlorides', 'total sulfur dioxide',
       'pH', 'sulphates', 'alcohol', 'quality']]
y= df.loc[:,'quality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
t= DecisionTreeRegressor(max_depth=5,min_samples_leaf= 20)
```

```python
t.fit(xtrain,ytrain)
```

```python
#Machine Learning Prediction Score
t.score(xtest,ytest)
```
