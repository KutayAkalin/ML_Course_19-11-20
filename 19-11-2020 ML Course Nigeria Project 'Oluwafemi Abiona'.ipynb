```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score 
```

```python
## Training
X  = wine.drop(['quality'], axis = 1)
y  = wine['quality']
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y,random_state=101)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# ML Application

- Define models.
- Fit models.
- Evaluate models for both train and test dataset.
- Generate Confusion Matrix and scores of Accuracy, Recall, Precision and F1-Score.
- Analyse occurrence of overfitting and underfitting. If there is any of them, try to overcome it within a different section.

```python
#The functions below will be used to measure the accuracy of the model

def generateClassificationReport_Tr(y_true,y_pred):
    '''Train data accuracy tester'''
    print(classification_report(y_true,y_pred));
    print(confusion_matrix(y_true,y_pred));
    print('\nTrain Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%');
    print('\nTrain Precision:',
          round(100*precision_score(y_true,y_pred),3),'%')
    print('\nTrain Recall:',
          round(100*recall_score(y_true,y_pred),3),'%')
    print('\nTrain F1 Score:',
          round(100*f1_score(y_true,y_pred),3),'%')
   
def generateClassificationReport_T(y_true,y_pred):
    '''Test data accuracy tester'''
    print(classification_report(y_true,y_pred));
    print(confusion_matrix(y_true,y_pred));
    print('\nTest Accuracy is: ',
          round(100*accuracy_score(y_true,y_pred),3),'%');
    print('\nTest Precision:',
          round(100*precision_score(y_true,y_pred),3),'%')
    print('\nTest Recall:',
          round(100*recall_score(y_true,y_pred),3),'%')
    print('\nTest F1 Score:',
          round(100*f1_score(y_true,y_pred),3),'%')               
```

### Baseline Model - Decision Tree

```python
#DECISION TREE
dc = DecisionTreeClassifier(max_depth=6, random_state=42);
dc.fit(X_train,y_train);
```

```python
###TRAIN DATA
ytrain_pred = dc.predict(X_train)
generateClassificationReport_Tr(y_train,ytrain_pred)
```

```python
###TEST DATA
ytest_pred = dc.predict(X_test)
generateClassificationReport_T(y_test,ytest_pred)
```

```python
#Feature Importance
plt.figure(figsize=(12, 8))
importance = dc.feature_importances_
sns.barplot(x=importance, y=X.columns)
plt.show()
```

### Model 2 - Logistic Regression

```python
#LOGISTIC REGRESSION
lr = LogisticRegression();
lr.fit(X_train,y_train);
```

```python
###TRAIN DATA
ytrain_pred = lr.predict(X_train)
generateClassificationReport_Tr(y_train,ytrain_pred)
```

```python
###TEST DATA
ytest_pred = lr.predict(X_test)
generateClassificationReport_T(y_test,ytest_pred)
```

### Model 3 - Support Vector Machine

```python
#SUPPORT VECTOR MACHINE
svc = SVC();
svc.fit(X_train,y_train);
```

```python
#TRAIN DATA
ytrain_pred = svc.predict(X_train)
generateClassificationReport_Tr(y_train,ytrain_pred)
```

```python
#TEST DATA
ytest_pred = svc.predict(X_test)
generateClassificationReport_T(y_test,ytest_pred)
```

### Model 4 - RandomForestClassifier

```python
#RANDOM FOREST
rfc = RandomForestClassifier(n_estimators=500)
rfc.fit(X_train,y_train);
```

```python
#TRAIN DATA
ytrain_pred = rfc.predict(X_train)
generateClassificationReport_Tr(y_train,ytrain_pred)
```

```python
#TEST DATA
ytest_pred = rfc.predict(X_test)
generateClassificationReport_T(y_test,ytest_pred)
```
